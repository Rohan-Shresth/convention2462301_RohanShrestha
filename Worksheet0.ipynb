```python
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

low = []
moderate = []
high = []

for study, ent, sleep in time_data:
    if study < 3:
        low.append(study)
    elif 3 <= study <= 5:
        moderate.append(study)
    else:
        high.append(study)

print("Low study hours:", low)
print("Moderate study hours:", moderate)
print("High study hours:", high)

print("Days with LOW study time:", len(low))
print("Days with MODERATE study time:", len(moderate))
print("Days with HIGH study time:", len(high))

study_minutes = []
for study, ent, sleep in time_data:
    minutes = study * 60
    study_minutes.append(minutes)

print("Study minutes:", study_minutes)

study_hours = []
entertainment_hours = []
sleep_hours = []

for study, ent, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(ent)
    sleep_hours.append(sleep)

avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)

print("Average Study Hours:", avg_study)
print("Average Entertainment Hours:", avg_entertainment)
print("Average Sleep Hours:", avg_sleep)

import matplotlib.pyplot as plt

plt.scatter(study_hours, sleep_hours, color="blue")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Pattern")
plt.show()
```

```python
# --------------------------------------------
# Task 1: Sum of Nested Lists (Recursion)
# --------------------------------------------

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)   # call function again (recursion)
        else:
            total += element
    return total


# --------------------------------------------
# Task 2: Generate All Permutations of a String
# --------------------------------------------

def generate_permutations(s):
    if len(s) == 0:
        return [""]

    if len(s) == 1:
        return [s]

    result = []

    for i in range(len(s)):
        letter = s[i]
        remaining = s[:i] + s[i+1:]   # remove chosen letter

        small_perms = generate_permutations(remaining)

        for perm in small_perms:
            result.append(letter + perm)

    # remove duplicates if any
    unique_result = []
    for p in result:
        if p not in unique_result:
            unique_result.append(p)

    return unique_result


# --------------------------------------------
# Task 3: Directory Size Calculation (Recursion)
# --------------------------------------------

def calculate_directory_size(directory):
    total = 0

    for key in directory:
        value = directory[key]

        if isinstance(value, dict):
            total += calculate_directory_size(value)  # recursive call
        else:
            total += value

    return total


# --------------------------------------------
# Testing All Tasks in One Program
# --------------------------------------------

# Task 1 Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Task 1: Sum of Nested List =", sum_nested_list(nested_list))

# Task 2 Test
print("Task 2: Permutations of 'abc' =", generate_permutations("abc"))
print("Task 2: Permutations of 'aab' =", generate_permutations("aab"))

# Directory structure for Task 3
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Task 3 Test
print("Task 3: Total Directory Size =", calculate_directory_size(directory_structure), "KB")
```
