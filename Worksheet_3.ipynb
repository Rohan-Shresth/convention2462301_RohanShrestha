```python
embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix='Embarked')
titanic = pd.concat([titanic, embarked_dummies], axis=1)

titanic = titanic.drop(columns=['Embarked'])
titanic.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic.groupby("Sex")["Survived"].mean()

sns.barplot(x="Sex", y="Survived", data=titanic)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower()

print("Columns after cleaning:", df.columns)

# Check for required columns
required = ['age', 'pclass']
for col in required:
    if col not in df.columns:
        raise KeyError(f"Missing required column: {col}")

# Identify survived column automatically
survive_col = None
for col in df.columns:
    if 'surv' in col:   # matches 'survived', 'survival', 'surv'
        survive_col = col
        break

if survive_col is None:
    raise KeyError("No column found for survival (e.g., survived, surv)")

print("Using survival column:", survive_col)

# Create age groups
df['agegroup'] = pd.cut(
    df['age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)

# Pivot
pivot = df.pivot_table(
    values=survive_col,
    index='agegroup',
    columns='pclass',
    aggfunc='mean'
)

# Plot
sns.heatmap(pivot, annot=True, cmap='Blues')
plt.title("Survival Rates by Age Group & Class")
plt.show()
```
